####**Le cas de base**

Le principe de la fonction récursive, en pratique, dépend d'un rappel de la même fonction jusqu'à ce qu'une condition spécifique soit atteinte. Nous appellons l'environnement de l'appel qui atteint cette condition "**le cas de base**". Les cas de bases sont conventionnellement programmés au début de la fonction afin de pouvoir les identifier facilement en tant que programmeur.

Par exemple: la fonction factorielle prend un nombre entier x et fait des appels avec (x-1), (x-2), etc. jusqu'à son **cas de base** x=0.


```
        if (x==0):
            return 1
```



####**L'étape récursive**

La majorité du temps, la partie compliquée est de programmer **l'étape récursive**. 

La chose importante qu'il faut garder en tête lorsque nous programmons l'étape récursive, c'est qu'il faut présumer que nous savons comment résoudre n'importe quelle récursion "plus petite". Cette présomption s'avère exacte en fait:

>Notre fonction récursive se rappelle elle-même, toujours plus creux dans la récursion sans rien résoudre, jusqu'à ce qu'elle appelle le **cas de base**. Nous savons comment résoudre le cas de base et nous l'avons encodé. Nos appels récursifs commencent à se résoudre du plus creux (le dernier appel utilise la solution rapportée par le cas de base) au plus près (chaque appel utilise la solution d'un appel plus profond maintenant résolu). C'est un principe très similaire à l'induction mathématique qui est en jeu. ##Mathématiques simple pour se pratiquer

Les fonctions récursives sont très semblables aux relations de récurrences en mathématique.

**Définition récursivité [Informatique]:** Une fonction récursive est une fonction qui, durant son exécution, fait appelle à elle-même.

**Définition récurrence [Mathématique]:** Une suite définie par récurrence est une suite définie par son (ou ses) premier(s) terme(s) et par une relation de récurrence, qui définit chaque terme à partir du précédent ou des précédents lorsqu'ils existent. ([lien wiki](https://fr.wikipedia.org/wiki/Suite_d%C3%A9finie_par_r%C3%A9currence))

>🙃Définition récursive de la récursivité: Une fonction récursive est une fonction qui, durant son exécution, fait un appel récursif.🙃 À NE PAS PRENDRE AU SÉRIEUX

Cette relation étroite entre la récurrence et la récursivité nous permet d'aiguiser nos dents de programmeurs en programmant des **fonctions récursives** qui représentent ces **suites définies par récurrences**. Nous avons déjà vu deux exemples célèbres dans les notes de cours sur la récursivité, soit la fonction factorielle **X! = X (X-1)!** et la suite Fibonnacci **Fib(n) = Fib(n-1) + Fib(n-2)**.

# Exemple 3: Le coefficients binomiaux

Le coefficient binomial **C(n,k)** représente le nombre de manières différentes de choisir k éléments parmis n éléments.

**C(n,k)** peut aussi être dénoté:

Binomial recurrence.PNG

Nous notons la relation de récurrence: **C(n,k) = C(n-1,k) + C(n-1,k-1)**

Nous notons aussi quelques cas spéciaux (lire: **cas de bases**):

**C(n,n) =  1**, puisqu'il n'y a qu'une seule manière de choisir n éléments parmi une sélection de n éléments (soit en les choisissant tous).

**C(n,0) =  1**, puisqu'il n'y a qu'une seule manière de choisir 0 éléments parmis une sélection de n éléments (soit en choisissant aucun).

Prenons par exemple une main initiale de poker Texas Hold'em qui est déterminé en distribuant 2 cartes d'un paquet de 52 cartes. Un nombre qui pourrait interreser un statisticien (ou un joueur) serait le nombre de mains différentes que l'on peut recevoir avec ses deux cartes. Ceci correspond exactement à **C(52,2)**, soit le nombre de manières différentes dont on peut recevoir 2 cartes d'un paquet de 52.

(Il est à noté ici que l'on considère recevoir [AS de COEUR, ROI de PIQUE] différent que de recevoir [AS de COEUR, ROI de TRÈFLE])

Voici comment se déroule le calcul de la récurrence: **C(4,2)**

= **C(3,2)** + **C(3,1)** 

= *C(2,2)* + **C(2,1)** + **C(2,1)** + *C(2,0)*

= *C(2,2)* + *C(1,1)* + *C(2,0)* + *C(2,0)* + *C(1,0)* + *C(2,0)*

= 1+1+1+1+1+1 = 6

###**Extra:**

> L'origine de la formule récursive nous provient d'une décomposition astucieuse du problème original. Pour calculer **C(n,k)** (le nombre de manières différentes de choisir k éléments parmi n), on compte deux évènements mutuellement exclusifs qui, lorsque combinés ensemble, forment la manière de compter **C(n,k)**. En additionnant ces deux comptes, nous obtenons l'égalité vue plus haut dans notre formule récursive **C(n,k) = C(n-1,k) + C(n-1,k-1)**

>>En Premier, nous comptons le nombre de manières différentes que nous pouvons choisir k éléments parmi les n **sans choisir le dernier élément**. Notons qu'ici nous ne faisons que choisir k éléments parmi n-1 éléments. Ceci représente la première partie de notre somme **C(n-1,k)**.

>>En deuxième, nous comptons le reste, soit le nombre de manières différentes de choisir k éléments parmi n **en ayant choisi le dernier élément**. Notons qu'ici on choisi k-1 éléments (puisque le dernier élément n'est pas vraiment un choix) parmi une liste de n-1 éléments (puisque nous ne pouvons pas choisir le dernier élément deux fois). Ceci représente la deuxième partie de notre somme **C(n-1,k-1)**

###**Extra 2:**

> En fait il existe un truc secret pour connaître un coefficient binomial, mais il implique 3 fonctions factorielles: **C(n,k) = n! / (k!(n-k)!)**
